```python
import simpy
import random
```

SE PARAMETRIZA LOS VALORES DE LA INSTITUCION

```python
# Solo se va a tener en cuenta una escuela de Nabon.
#Cantidad de estudiantes de la institucion
AULAS = 15
CANTIDAD_ESTUDIANTES = 140
#Cantidad de docentes de la institucion.
CANTIDAD_DOCENTES = 29
print("Cantidad de docentes",CANTIDAD_DOCENTES," Cantidad de estudiantes ",CANTIDAD_ESTUDIANTES)

#Se obtiene el %90 de docentes vacunados
V_DOCENTES = int((CANTIDAD_DOCENTES*90)/100)
print("El numero de docentes vacunados: {}".format(V_DOCENTES ))

#Se obtiene un numero de 5 al 10  para sacar un porcentaje de estudiante no vacunados
aleatorio = random.randint(5, 10)
NO_V_ESTUDIANTES = int((CANTIDAD_ESTUDIANTES*aleatorio)/100)
V_ESTUDIANTES = CANTIDAD_ESTUDIANTES - NO_V_ESTUDIANTES
print("El numero de estudiantes vacunados: {}".format(V_ESTUDIANTES))
print(NO_V_ESTUDIANTES)

#Se crea la agregacion de estudiantes por aula
es_x_aula = round(CANTIDAD_ESTUDIANTES / AULAS)
print("Estudiantes ",es_x_aula,"por aula ",AULAS)

```

SE CREA LOS METODOS NECESARIOS PARA LA INTERACION DE LOS ESTUDIANTES

```python
#Se crea la clase aula
#Se parametriza el tiempo de clase,cantidad de clases por dia,tiempo reseco de 30

class Escuela():
  def __init__(self, env, num_aulas, nombre_institucion):
    self.env = env
    self.aulas = simpy.Resource(env, num_aulas)

def llenado(env, estudiante, colegio):
  with colegio.aulas.request() as aula:
    if estudiante < es_x_aula - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 1'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*2) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 2'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*3) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 3'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*4) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 4'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*5) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 5'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*6) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 6'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*7) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 7'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*8) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 8'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*9) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 9'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*10) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 10'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*11) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 11'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*12) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 12'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*13) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 13'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*14) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 14'.format(estudiante, env.now))
      yield aula
    elif estudiante < (es_x_aula*15) - 1:
      print('Entra el estudiante {} en el tiempo {} al aula 15'.format(estudiante, env.now))
      yield aula
    
#Metodo para crear     
def llenado_doc(env, docente, colegio, au):
   with colegio.aulas.request() as aula:   
    if docente>0:
      print("El docente {} llega en el tiempo {} al aula {}".format(docente, env.now, au + 1)) if au < 15 else None
      yield aula
        
def inicio_clases(env, docente, colegio, au):
   with colegio.aulas.request() as aula:   
    if docente>0:
      print("El docente {} inicia la clase en el tiempo {} al aula {}".format(docente, env.now, au + 1)) if au < 15 else None
      yield aula

def fin_primeras_horas(env, docente, colegio, au):
  with colegio.aulas.request() as aula:
    if docente>0:
      print("El docente {} termino la clase en el tiempo {} en el aula {}".format(docente, env.now, au + 1)) if au < 15 else None
      yield aula
        
def salida_docente_primeras_horas(env, colegio, docente, au):
  with colegio.aulas.request() as aula:
    print("El docente {} sale en el tiempo {} de la aula {}".format(docente, env.now, au + 1)) if au < 15 else None
    yield aula
def entra_docente_segundas_horas(env, colegio, docente_nuevo, au):
  with colegio.aulas.request() as aula:
    print("Cambio de aulas para los docentes")
    print("El docente {} entra en el tiempo {} a la aula {}".format(docente_nuevo, env.now, au + 1)) if au < 15 else None
    yield aula
    
def inicio_segundas_horas(env, colegio, docente_nuevo, au):
  with colegio.aulas.request() as aula:
    print("El docente {} inicia la clase en el tiempo {} al aula {}".format(docente_nuevo, env.now, au + 1)) if au < 15 else None
    yield aula

def fin_segundas_horas(env, colegio, docente_nuevo, au):
  with colegio.aulas.request() as aula:
    print("El docente {} termino la clase en el tiempo {} en el aula {}".format(docente_nuevo, env.now, au + 1)) if au < 15 else None
    yield aula

def salida_docente_segundas_horas(env, colegio, docente_nuevo, au):
  with colegio.aulas.request() as aula:
    print("El docente {} sale en el tiempo {} de la aula {}".format(docente_nuevo, env.now, au + 1)) if au < 15 else None
    yield aula

def salida_alumnos_recreo(env, estudiante, colegio):
  with colegio.aulas.request() as aula:
    print("El estudiante {} sale al recreo en el tiempo {} por 30 minutos".format(estudiante, env.now))
    yield aula 
    
def salida_alumnos_recreo(env, estudiante, colegio):
  with colegio.aulas.request() as aula:
    print("El estudiante {} sale al recreo en el tiempo {} por 30 minutos".format(estudiante, env.now))
    yield aula

def salida_alumnos_casa(env, estudiante, colegio):
  with colegio.aulas.request() as aula:
    print("El estudiante {} sale en el tiempo {} a su casa".format(estudiante, env.now))
    yield aula
```

```python
def simulacion(env, num_aulas, estudiantes, docentes):
  institucion = Escuela(env, num_aulas, "FRANCISCO TAMARIZ")
  for di in range(31):
    if (di + 1) < 31:
      print("-----------------LLEGADA DE LOS ALUMNOS A LAS AULAS DIA {} -----------------".format(di + 1))
      for i in range(estudiantes):
            env.process(llenado(env,(i+1),institucion))
      yield env.timeout(2)

      print("***LLEGADA DE LOS DOCENTES A LAS AULAS***")
      d = list(range(docentes))
      a = random.sample(d, 15)
      b = random.sample(d, 15)
      c = random.sample(d, 15)
      d = random.sample(d, 15)
      print(d,a,b,c,d)
      for j in range(num_aulas):
        env.process(llenado_doc(env, a[j]+ 1, institucion, j))
      yield env.timeout(1)

      print("***INICIO DE LAS CLASES DE LOS DOCENTES EN LAS AULAS***")
      for k in range(num_aulas):
        env.process(inicio_clases(env, a[k]+ 1, institucion, k))
      yield env.timeout(82) # 41 equivale a una hora de clase como son 2 por clase seria 82

      print("***FIN DE LAS PRIMERAS 2 HORAS DE CLASES DE LOS DOCENTES EN LAS AULAS***")
      for l in range(num_aulas):
        env.process(fin_primeras_horas(env, a[l] + 1, institucion, l))
      yield env.timeout(1)

      print("***SALIDAS DEL DOCENTE DE LA AULAS ***")
      for m in range(num_aulas):
        env.process(salida_docente_primeras_horas(env, institucion, a[m] + 1, m))
      yield env.timeout(0.3)

      print("***CAMBIO Y ENTRADA DEL DOCENTE A OTRAS AULAS PARA CLASES ANTES DEL RECREO***")
      for n in range(num_aulas):
        env.process(entra_docente_segundas_horas(env, institucion, b[n] + 1, n))
      yield env.timeout(1)

      print("***INICIO DE LAS SEGUNDAS HORAS DE CLASES DE LOS DOCENTES EN LAS AULAS ANTES DEL RECREO***")
      for o in range(num_aulas):
        env.process(inicio_segundas_horas(env, institucion, b[o] +1, o))
      yield env.timeout(82) # 41 equivale a una hora de clase como son 2 por clase seria 82

      print("***TERMINO DE LAS SEGUNDAS HORAS DE CLASES DE LOS DOCENTES EN LAS AULAS ANTES DEL RECREO***")
      for p in range(num_aulas):
        env.process(fin_segundas_horas(env, institucion, b[p] +1, p))
      yield env.timeout(0.5)

      print("***SALIDAS DEL DOCENTE DE LA AULAS PAL RECESO ***")
      for q in range(num_aulas):
        env.process(salida_docente_segundas_horas(env, institucion, b[q] + 1, q))
      yield env.timeout(0.2)

      print("***SALIDA DE LOS ESTUDIANTES AL RECREO***")
      for r in range(estudiantes):
        env.process(salida_alumnos_recreo(env, (r+1), institucion))
      yield env.timeout(30) # Duracion del receso como 30 minutos

      print("***ENTRADA DE LOS ESTUDIANTES DESPUES DEL RECREO***")
      for i in range(estudiantes):
        env.process(llenado(env,(i+1),institucion))
      yield env.timeout(1)
      
      print("***LLEGADA DE LOS DOCENTES A LAS AULAS DESPUES DEL RECESO***")
      for j in range(num_aulas):
        env.process(llenado_doc(env, c[j]+ 1, institucion, j))
      yield env.timeout(2)

      print("***INICIO DE LAS CLASES DE LOS DOCENTES EN LAS AULAS DESPUES DEL RECESO***")
      for k in range(num_aulas):
        env.process(inicio_clases(env, c[k]+ 1, institucion, k))
      yield env.timeout(82) # 60 equivale a un minuto (60 * 82)

      print("***FIN DE LAS TERCERAS 2 HORAS DE CLASES DE LOS DOCENTES EN LAS AULAS DESPUES DEL RECESO***")
      for l in range(num_aulas):
        env.process(fin_primeras_horas(env, c[l] + 1, institucion, l))
      yield env.timeout(1)

      print("***SALIDAS DEL DOCENTE DE LA AULAS DESPUES DEL RECESO***")
      for m in range(num_aulas):
        env.process(salida_docente_primeras_horas(env, institucion, c[m] + 1, m))
      yield env.timeout(0.3)

      print("***CAMBIO Y ENTRADA DEL DOCENTE A OTRAS AULAS PARA CLASES DESPUES DEL RECESO***")
      for n in range(num_aulas):
        env.process(entra_docente_segundas_horas(env, institucion, d[n] + 1, n))
      yield env.timeout(1)

      print("***INICIO DE LAS SEGUNDAS HORAS DE CLASES DE LOS DOCENTES EN LAS AULAS DESPUES DEL RECESO***")
      for o in range(num_aulas):
        env.process(inicio_segundas_horas(env, institucion, d[o] +1, o))
      yield env.timeout(82) # 41 equivale a una hora de clase como son 2 por clase seria 82

      print("***TERMINO DE LAS SEGUNDAS HORAS DE CLASES DE LOS DOCENTES EN LAS AULAS DESPUES DEL RECREO***")
      for p in range(num_aulas):
        env.process(fin_segundas_horas(env, institucion, d[p] +1, p))
      yield env.timeout(0.5)

      print("***SALIDA DE LOS ESTUDIANTES A SUS CSAS***")
      for r in range(estudiantes):
        env.process(salida_alumnos_casa(env, (r+1), institucion))
      yield env.timeout(0) 
    elif (di + 1) == 31:
      alumnos_pcr = round((estudiantes*10)/100)
      print("DESPUES DE 30 DIAS SE HARA PRUEBAS AL 10% DE LOS ESTUDIANTES: {}".format(alumnos_pcr))
      print("DE ESAS PRUEBAS AL AZAR DEL 10% DE LOS ESTUDIANTES, ESTA CANTIDAD DIO POSTIVO PARA COVID-19: {}".format(round((alumnos_pcr*2)/100)))

if __name__ == "__main__":
  print("Escuela en Cuenca")
  # Creamos el entorno de simulacion
  env=simpy.Environment()
  print("# de Estudinates vacunados  => ", V_ESTUDIANTES)
  env.process(simulacion(env, AULAS, V_ESTUDIANTES, V_DOCENTES))
  # Ejecutamos el proceso durante el tiempo de simulacion
  env.run(until = 1000)
```
